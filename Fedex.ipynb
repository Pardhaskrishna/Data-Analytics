```python
shipment_mode_mode = df['Shipment Mode'].mode()[0]
df['Shipment Mode'].fillna(shipment_mode_mode, inplace=True)

df['Line Item Insurance (USD'] = df['Line Item Insurance (USD)'].fillna(0)

df.drop_duplicates(inplace=True)

date_columns = ['PQ First Sent to Client Date', 'PO Sent to Vendor Date', 
                'Scheduled Delivery Date', 'Delivered to Client Date', 
                'Delivery Recorded Date']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')
    

numeric_columns = ['Line Item Quantity', 'Line Item Value', 'Weight (Kilograms)', 
                   'Freight Cost (USD)', 'Line Item Insurance (USD)']
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df['Country'] = df['Country'].str.strip().str.lower()  # Example normalization

# Check the cleaned data
print(df.info())
```
